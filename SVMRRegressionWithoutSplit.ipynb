```python
#import pandas
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
train_independent=sc.fit_transform(independent)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

param_grid={'kernel':['linear','rbf','poly'],
           'gamma':['scale','auto'],
           'C':[5,10]}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(train_independent,dependent)
```

```python
re=grid.cv_results_
print(re)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid.best_params_
```

```python
Y_predict=grid.predict(train_independent)
Y_predict
```

```python
from sklearn.metrics import r2_score
r2=r2_score(dependent,Y_predict)
r2
```

```python
preinput=sc.transform([[40,28.5,2,1,0]])
result=grid.predict(preinput)
result
```
