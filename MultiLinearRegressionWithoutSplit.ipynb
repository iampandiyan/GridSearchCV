```python
#import pandas library for the data processing
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
#Print the columns in the csv
dataset.columns
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
#get_dummies with drop_first=True to remove the repedetive column
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
#print the column names
dataset.columns
```

```python
#Prepare the dependent/input dataset
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
#prepare the dependent/output dataset
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
train_independent=sc.fit_transform(independent)
```

```python
#Create Model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

param_grid={'fit_intercept':[True,False],
           'copy_X':[True,False],
           'n_jobs':[None,5,10]}
grid=GridSearchCV(LinearRegression(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(train_independent,dependent)
```

```python
re=grid.cv_results_
print(re)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid.best_params_
```

```python
#result of the predict X_test
Y_predict=grid.predict(train_independent)
Y_predict
```

```python
#Validation of the model Y_predict Vs Y_test
from sklearn.metrics import r2_score
r2=r2_score(dependent,Y_predict)
r2
```

```python
preinput=sc.transform([[40,28.5,2,1,0]])
result=grid.predict(preinput)
result
```
